```python
# Importamos datos de hipotecas a partir de un csv

# 1) Rutas del ficheros 

RUTA_BASE = r"C:/Users/Elías/Desktop/Bootcamp/Proyectos/Week 4"
RUTA_FICHERO   = rf"{RUTA_BASE}/Hipotecas_ipc.csv"

# 2) Cargar hipotecas

df_hipotecas_ipc = pd.read_csv(RUTA_FICHERO, low_memory=False, encoding="utf-8")
```

```python
df_hipotecas_ipc.head(5)
```

```python
df_hipotecas = df_hipotecas_ipc[["comunidad", "year", "num_hipotecas"]].copy()
```

```python
df_hipotecas = df_hipotecas.rename(columns={
    "year": "anio",
    "num_hipotecas": "n_hipotecas"
})

df_hipotecas.head(5)
```

```python
# 2) Rutas a los ficheros 

RUTA_BASE = r"C:/Users/Elías/Desktop/Bootcamp/Proyectos/Week 4"
RUTA_SALARIOS   = rf"{RUTA_BASE}/Salarios.xls"


# 3) Cargar excel

df_salarios = pd.read_excel(RUTA_SALARIOS, engine="xlrd") 

df_salarios.head(5)
```

```python
df_hipotecas["anio"] = pd.to_datetime(df_hipotecas["anio"])
df_hipotecas = df_hipotecas[df_hipotecas["anio"].dt.month == 1]
```

```python
map_df_hipotecas = {
    'andalucia': 'Andalucia',
    'aragon': 'Aragon',
    'cantabria': 'Cantabria',
    'castilla-leon' : 'Castilla y Leon',
    'castilla-mancha': 'Castilla La Mancha',
    'catalunya': 'Catalunya',
    'ceuta': 'Ceuta',
    'comunidad-foral-navarra': 'Navarra',
    'comunidad-madrid': 'Madrid',
    'comunitat-valenciana': 'Comunitat Valenciana',
    'extremadura': 'Extremadura',
    'galicia': 'Galicia',
    'islas-baleares': 'Islas Baleares',
    'Canarias': 'Canarias',
    'la-rioja': 'La Rioja',
    'pais-vasco': 'Pais Vasco',
    'region-murcia': 'Murcia',
    'principado-asturias': 'Asturias',
    'melilla': 'Melilla'
}
df_hipotecas["comunidad"] = df_hipotecas["comunidad"].replace(map_df_hipotecas) 

df_hipotecas.head(5)
```

```python
df_hipotecas['anio'] = pd.to_datetime(df_hipotecas['anio'], errors='coerce').dt.year
```

```python
df_hipotecas['anio'] = pd.to_numeric(df_hipotecas['anio'], errors='coerce')
```
